```python
import os
from pathlib import Path

# Set custom cache directory to local ./data folder
os.environ['KAGGLEHUB_HOME'] = str(Path.cwd() / 'data')

# Download latest version of alternative dataset
path = kagglehub.dataset_download("xhlulu/140k-real-and-fake-faces")

print("Path to dataset files:", path)
print("Download destination set to:", os.environ['KAGGLEHUB_HOME'])
```

```python
import shutil
import random
from pathlib import Path

# After downloading, organize files into train/val/test splits
# This splits the downloaded data into 70% train, 15% val, 15% test

source_root = Path('./data/datasets/philosopher0808/real-vs-ai-generated-faces-dataset/versions/1/data_source/data_source')
dest_root = Path('./data/organized')  # Final organized structure

# Create directory structure
for split in ['train', 'val', 'test']:
    for cls in ['real', 'fake']:
        (dest_root / split / cls).mkdir(parents=True, exist_ok=True)

# Get all files
real_files = list(source_root.glob('ffhq/*.jpg'))
fake_files = list(source_root.glob('fake/*.jpg'))

print(f"Found {len(real_files)} real images and {len(fake_files)} fake images")

# Shuffle and split (70% train, 15% val, 15% test)
random.seed(42)
random.shuffle(real_files)
random.shuffle(fake_files)

def split_dataset(files):
    n = len(files)
    train_idx = int(0.7 * n)
    val_idx = int(0.85 * n)
    return {
        'train': files[:train_idx],
        'val': files[train_idx:val_idx],
        'test': files[val_idx:]
    }

real_split = split_dataset(real_files)
fake_split = split_dataset(fake_files)

# Copy files
for split in ['train', 'val', 'test']:
    for f in real_split[split]:
        shutil.copy2(f, dest_root / split / 'real' / f.name)
    for f in fake_split[split]:
        shutil.copy2(f, dest_root / split / 'fake' / f.name)

print("\n✓ Dataset organized successfully!")
print(f"  Train: {len(real_split['train']) + len(fake_split['train'])} images")
print(f"  Val: {len(real_split['val']) + len(fake_split['val'])} images")  
print(f"  Test: {len(real_split['test']) + len(fake_split['test'])} images")
print(f"\nReady to train! Update your notebook to use: {dest_root}")
```
